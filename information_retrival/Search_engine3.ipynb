```python
df=pd.read_csv('/content/finaltokenized.csv')
```

```python
import pyterrier as pt
if not pt.started():
    pt.init()
```

```python
text_columns=['official_title', 'detailed_description', 'eligibility_inclusion_criteria', 'eligibility_exclusion_criteria', 'eligibility_gender', 'eligibility_minimum_age', 'eligibility_maximum_age', 'eligibility_healthy_volunteers']

df['combined_text'] = df[text_columns].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

df['docno'] = df['nct_id']

indexer = pt.DFIndexer("./index_sampledocs", overwrite=True, stemmer= False, stopwords=False, pretokenised = True)
indexref = indexer.index(df['combined_text'], df['docno'])
```

```python
index = pt.IndexFactory.of(indexref)
```

```python
# Queries
!pip install --upgrade ir_datasets
dataset = pt.get_dataset('irds:clinicaltrials/2021/trec-ct-2022')
queries = dataset.get_topics(variant='text')
#This is a dataframe that contains the queries
display(queries)
```

```python
br = pt.BatchRetrieve(index, wmodel="Tf") # TF_IDF, BM25, PL2
#As I am using a single query (that I manually provide I use search)

br.transform(queries)
```

```python
#Qrels

path_to_qrels = 'clinical_qrels22.txt'
qrels = pd.read_csv(path_to_qrels, names=['qid','Q0','docno','label'],sep=" ",header=None)
qrels = qrels.drop(columns=['Q0'])
qrels["qid"] = qrels["qid"].astype(str)
qrels["docno"] = qrels["docno"].astype(str)
display(qrels.head())
```

```python
from pyterrier.measures import *
MODEL = pt.BatchRetrieve(index, wmodel="BM25") # TF_IDF, BM25

results = pt.Experiment(
    [MODEL],
    queries,
    qrels,
    eval_metrics=[P@5,P@10])
```
